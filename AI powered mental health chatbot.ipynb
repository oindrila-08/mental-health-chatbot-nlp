```python
!pip install nltk scikit-learn joblib

import pandas as pd
import string
import joblib
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    tokens = nltk.word_tokenize(text.lower())
    tokens = [t for t in tokens if t not in string.punctuation]
    tokens = [lemmatizer.lemmatize(t) for t in tokens if t not in stop_words]
    return ' '.join(tokens)
```

```python
train_df = pd.read_csv("/content/train.csv", names=['text_label'])
test_df = pd.read_csv("/content/test.csv", names=['text_label'])
val_df = pd.read_csv("/content/val.csv", names=['text_label'])

# Split text and label
train_df[['text', 'label']] = train_df['text_label'].str.split(';', expand=True)
test_df[['text', 'label']] = test_df['text_label'].str.split(';', expand=True)
val_df[['text', 'label']] = val_df['text_label'].str.split(';', expand=True)

# Clean text data
train_df['cleaned'] = train_df['text'].apply(clean_text)
test_df['cleaned'] = test_df['text'].apply(clean_text)
val_df['cleaned'] = val_df['text'].apply(clean_text)

train_df.head()
```

```python
# Create a pipeline
model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB()),
])

# Train the model
model.fit(train_df['cleaned'], train_df['label'])

# Predict on the test data
predictions = model.predict(test_df['cleaned'])

# Evaluate the model
print(classification_report(test_df['label'], predictions))
print("Accuracy:", accuracy_score(test_df['label'], predictions))
```

```python
X_train = train_df['cleaned']
y_train = train_df['label']

X_test = test_df['cleaned']
y_test = test_df['label']

model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('nb', MultinomialNB())
])

model.fit(X_train, y_train)
joblib.dump(model, "chatbot_model.pkl")
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'tfidf__ngram_range': [(1, 1), (1, 2)],
    'tfidf__use_idf': [True, False],
    'tfidf__smooth_idf': [True, False],
    'nb__alpha': [0.1, 0.5, 1.0],
}
```

```python
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```
